### Reshape return  a new view and does not change original array while resize change orignal array

```python
a = np.array([1, 2, 3])
a[:, np.newaxis]
```

```python
a
```

```python
b = a.reshape(3,-1)
b[0, 0] = 99
print(a)
```

## mathematical operations

```python
a=np.array([[2,4,6],[8,10,12]])
b=np.array([[1,2,4],[4,5,6]])
```

```python
np.add(a,b)
```

```python
np.subtract(a,b)
```

```python
np.multiply(a,b)
```

```python
np.dot(a,b.T)
```

```python
a = np.array([1, 2,3])
b = np.array([4, 5, 6])
```

```python
arr=np.array([2,4,3,8,4])
print(f"mean : {np.mean(arr)}")
print(f" Square root {np.sqrt(arr)}")
print(f" median {np.median(arr)}")
print(f" standard devaition {np.std(arr)}")
```

```python
values, counts = np.unique(arr, return_counts=True)
mode = values[np.argmax(counts)]

print(f"mode {mode}")
```

## Broadcasting

### Broadcasting is used to perform mathematical operation on matix of different dimensions by raplicating element in lower dimention array to match to higher dimension array.

```python
a=np.array([[2,4,6],[6,8,10]])
b=np.array([2,4,6])
np.dot(a,b)
```
