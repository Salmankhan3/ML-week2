```python
#x['C']=x['C'].astype(str)
#x['S']=x['S'].astype(str)
#x['Q']=x['Q'].astype(str)
```

```python
x.columns = x.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
logmodel.coef_
```

```python
logmodel.intercept_
```

```python
logmodel.classes_
```

```python
logmodel.feature_names_in_
```

```python
prediction=logmodel.predict(X_test)
```

```python
from sklearn.metrics import f1_score, recall_score, accuracy_score,precision_score
```

```python
f1=f1_score(prediction,ytest)
recall=recall_score(prediction,ytest)
accuracy=accuracy_score(prediction,ytest)
precision=precision_score(prediction,ytest)
print(f'f1 score : {f1:.2f}%')
print(f'recall : {recall:.2f}%')
print(f'precision : {precision:.2f}%')
print(f'accuracy : {accuracy:.2f}%')
```
