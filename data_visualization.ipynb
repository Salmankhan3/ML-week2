```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("titanic_cleaned.csv")
```

```python
data
```

```python
ages = data['age']
```

```python
age_counts = ages.value_counts().sort_index()
```

## Age Destribution using line plot

```python
plt.figure(figsize=(9,5))
plt.plot(age_counts.index,age_counts.values,label='Age',marker="o",markeredgecolor='red',linewidth=2)
plt.title("Age Destribution",fontdict={'fontname':'Arial','fontsize':15,'color':'Blue'})
plt.xlabel("Age",fontdict={'fontname':'Arial','fontsize':16,'color':'Red'})
plt.ylabel("Age count",fontdict={'fontname':'Arial','fontsize':16,'color':'#893656'})
plt.legend()
plt.savefig('visualizations/age_destribution.jpeg')
plt.show()
```

```python
fare=data['fare']
plt.figure(figsize=(9,5))

plt.scatter(ages,fare,marker="o",linewidths=3,color="#192929")
plt.title("Age Vs Fare",fontdict={'fontname':'Arial','fontsize':15,'color':'Blue'})
plt.xlabel("Age",fontdict={'fontname':'Arial','fontsize':16,'color':'Red'})
plt.ylabel("Fare",fontdict={'fontname':'Arial','fontsize':16,'color':'#097802'})
plt.legend()
#plt.savefig('visualizations/age_vs_fare.jpeg')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.hist(data['pclass'],bins=25,align='left')
plt.title("Passenge class Destribution")
plt.xticks(data['pclass'])
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.savefig('visualizations/Pasenger_class_dest.png')
plt.show()
```

```python
first_class=data[data['pclass']==1]
len(first_class)
```

```python
survival_counts = data['survived'].value_counts().sort_index()
```

```python
survival_counts
```

```python
plt.bar(['Not Survived','Survived'],survival_counts.values,width=0.2)
plt.title("Survival Rate")
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.savefig('visualizations/survival_rate.png')
plt.legend()
plt.show()
```

```python
first_class=data[data['pclass']==1]['fare']
second_class=data[data['pclass']==2]['fare']
third_class=data[data['pclass']==3]['fare']
```

```python
plt.figure(figsize=(10,5))
plt.boxplot([first_class,second_class,third_class], 
        label=['1st Class', '2nd Class', '3rd Class'], patch_artist=True)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.grid(True)
plt.show()
```

```python
corr=data.select_dtypes(include='number').drop(['Unnamed: 0','Has cabin'],axis=1).corr()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True,fmt='.1f',cmap='Blues')
plt.savefig('visualizations/Corelation.png')
plt.show()
```

```python
age_gender = data[['age', 'sex']]
age_gender
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(x='sex', y='age', data=age_gender, palette='Pastel1')

plt.title("Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.grid(True)
plt.savefig('visualizations/voilin_plot.jpg')
plt.show()
```
